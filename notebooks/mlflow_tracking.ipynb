```python
import mlflow

with mlflow.start_run(run_name="Baseline_LightGBM"):

    # Model Parameters
    mlflow.log_param("model_type", "LightGBM")
    mlflow.log_param("n_estimators", 200)
    mlflow.log_param("learning_rate", 0.05)
    mlflow.log_param("num_leaves", 31)

    # Metrics
    mlflow.log_metric("RMSE", 1.8563)
    mlflow.log_metric("MAE", 0.8407)
    mlflow.log_metric("WAPE", 98.84)

    print("Baseline run logged successfully")
```

```python
with mlflow.start_run(run_name="Causal_LightGBM"):

    # Parameters
    mlflow.log_param("model_type", "LightGBM")
    mlflow.log_param("target", "causal_sales")
    mlflow.log_param("causal_method", "DoWhy_Backdoor_Adjustment")

    # Metrics
    mlflow.log_metric("RMSE", 1.8653)
    mlflow.log_metric("MAE", 0.8415)
    mlflow.log_metric("WAPE", 98.92)

    # Causal Metrics
    mlflow.log_metric("ATE", -0.055745695026705455)

    print("Causal run logged successfully")
```

```python
with mlflow.start_run(run_name="Causal_LightGBM"):

    mlflow.log_param("model_type", "LightGBM")
    mlflow.log_param("target", "causal_sales")

    mlflow.log_param(
        "causal_method",
        "DoWhy_Backdoor_Adjustment"
    )

    mlflow.log_metric("RMSE", 1.8653)
    mlflow.log_metric("MAE", 0.8415)
    mlflow.log_metric("WAPE", 98.92)

    mlflow.log_metric(
        "ATE",
        -0.055745695026705455
    )

    print("Causal run logged")
```
