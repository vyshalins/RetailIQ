```python
sales_long['date'] = pd.to_datetime(sales_long['date'])

sales_long = sales_long.sort_values(
    ['store_id', 'item_id', 'date']
)

sales_long['lag_7'] = (
    sales_long
    .groupby(['store_id', 'item_id'])['sales']
    .shift(7)
)

sales_long[
    ['date', 'sales', 'lag_7']
].head(15)
```

```python
sales_long['lag_14'] = (
    sales_long
    .groupby(['store_id', 'item_id'])['sales']
    .shift(14)
)

sales_long['lag_28'] = (
    sales_long
    .groupby(['store_id', 'item_id'])['sales']
    .shift(28)
)
```

```python
sample = sales_long[
    sales_long['item_id'] == sales_long['item_id'].iloc[0]
][['date', 'sales', 'lag_7', 'lag_14', 'lag_28']]

sample.head(35)
```

```python
sales_long['rolling_mean_7'] = (
    sales_long
    .groupby(['store_id', 'item_id'])['sales']
    .transform(
        lambda x: x.shift(1).rolling(7).mean()
    )
)

sales_long['rolling_mean_28'] = (
    sales_long
    .groupby(['store_id', 'item_id'])['sales']
    .transform(
        lambda x: x.shift(1).rolling(28).mean()
    )
)
```

```python
sample = sales_long[
    sales_long['item_id'] == sales_long['item_id'].iloc[0]
][[
    'date',
    'sales',
    'rolling_mean_7',
    'rolling_mean_28'
]]

sample.head(40)
```

```python
sales_long['price_lag_7'] = (
    sales_long
    .groupby(['store_id', 'item_id'])['sell_price']
    .shift(7)
)
```

```python
sales_long['price_change_pct'] = (
    (sales_long['sell_price'] - sales_long['price_lag_7'])
    / sales_long['price_lag_7']
)
```

```python
sales_long['discount_flag'] = (
    sales_long['price_change_pct'] < 0
).astype(int)
```

```python
validation_start = (
    model_df['date'].max() -
    pd.Timedelta(days=27)
)

train_df = model_df[
    model_df['date'] < validation_start
].copy()

valid_df = model_df[
    model_df['date'] >= validation_start
].copy()

print("Train Shape :", train_df.shape)
print("Valid Shape :", valid_df.shape)

print("\nTrain Date Range")
print(train_df['date'].min(), "to", train_df['date'].max())

print("\nValidation Date Range")
print(valid_df['date'].min(), "to", valid_df['date'].max())
```

```python
feature_cols = [
    # Demand history
    'lag_7',
    'lag_14',
    'lag_28',
    'rolling_mean_7',
    'rolling_mean_28',

    # Price features
    'sell_price',
    'price_change_pct',
    'discount_flag',

    # Calendar features
    'month',
    'year',
    'week_of_year',
    'is_weekend',

    # SNAP
    'snap_CA',
    'snap_TX',
    'snap_WI',

    # Events
    'has_event'
]

target_col = 'sales'

print("Number of Features:", len(feature_cols))
print(feature_cols)
```

```python
train_model = train_df.dropna(subset=feature_cols + [target_col])

valid_model = valid_df.dropna(subset=feature_cols + [target_col])

print("Train Shape:", train_model.shape)
print("Valid Shape:", valid_model.shape)
```

```python
X_train = train_model[feature_cols]
y_train = train_model[target_col]

X_valid = valid_model[feature_cols]
y_valid = valid_model[target_col]

print(X_train.shape)
print(X_valid.shape)
```

```python
from lightgbm import LGBMRegressor

lgb_model = LGBMRegressor(
    n_estimators=200,
    learning_rate=0.05,
    num_leaves=31,
    random_state=42,
    n_jobs=-1
)

lgb_model.fit(
    X_train,
    y_train
)
```
